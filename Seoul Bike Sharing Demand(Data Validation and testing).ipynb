## One sample T-test

One sample ttest is used to test whether there is difference in sample in population data

```python
import pandas as pd
import numpy as np
```

```python
data_frame=pd.read_csv('SeoulBikeData.csv',encoding='unicode_escape')
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
#import onesample ttest from scipy.stats
from scipy.stats import ttest_1samp
```

```python
def one_sample_test(data):
    h0='There is no difference'
    h1='There is a difference'
    data_length=len(data)
    mean=np.mean(data)
    sample_size=2500
    sample_data=np.random.choice(data,sample_size)
    t_test,p_value=ttest_1samp(sample_data,popmean=mean)
    print('t_test:',t_test)
    print('p_value:',p_value)
    if p_value<0.5:
        print('We accept alternate hypothesis h1')
        print(h1)
    else:
        print('We accept null hypothesis h0')
        print(h0)
```

```python
for cols in data_frame.columns:
    if data_frame[cols].dtype!='O':
        print('One Sample test for',cols)
        one_sample_test(data_frame[cols])
        print('--------------------------------------------------')
```

# Two-sample T-test with python

The independent samples t-test or 2-sample t-test compares the mean of two independent groups in order to determine whether there is a statistical evidence that the associated population means are significantly different. The independent samples t-test is a parametric test. This test is also known a: Independent t test. 

```python
from scipy.stats import ttest_ind
```

```python
def two_sample_ttest(data1,data2):
    h0='There is no difference'
    h1='There is a difference'
    ttest,p_value=ttest_ind(a=data1,b=data2,equal_var=False)
    print('t_test:',ttest)
    print('p_value:',p_value)
    if p_value<0.5:
        print('We accept alternate hypothesis h1')
        print(h1)
    else:
        print('We accept null hypothesis h0')
        print(h0)
```

```python
for col1 in data_frame.columns:
    for col2 in data_frame.columns:
        if (data_frame[col1].dtype!='O')&(data_frame[col2].dtype!='O'):
            if col1!=col2:
                print('Two sample ttest for {} and {}'.format(col1,col2))
                two_sample_ttest(data_frame[col1],data_frame[col2])
                print('-----------------------------------------------------------------------------')
```

# Paired t-test with Python. 
When you want to check how different samples the sample group are, you can go for a paired T-test.

```python
from scipy.stats import ttest_rel
```

```python
def paired_ttest(data1,data2):
    h0='There is no difference'
    h1='There is a difference'
    ttest,p_value=ttest_rel(a=data1,b=data2)
    print('t_test:',ttest)
    print('P value: ',p_value)
    if p_value<0.5:
        print('We accept alternate hypothesis h1')
        print(h1)
    else:
        print('We accept null hypothesis h0')
        print(h0)
```

```python
for col1 in data_frame.columns:
    for col2 in data_frame.columns:
        if (data_frame[col1].dtype!='O')&(data_frame[col2].dtype!='O'):
            if col1!=col2:
                print('Paired ttest for {} and {}'.format(col1,col2))
                paired_ttest(data_frame[col1],data_frame[col2])
                print('-----------------------------------------------------------------------------')
```

```python
def chi2_test(data1,data2):
    h0='There is difference'
    h1='There is no difference'
    alpha=0.05
    observed_values=pd.crosstab(data1,data2).values
    #step3: calculating the degree of freedom.
    values=chi2_contingency(observed_values)
    #so, we get chi-square value, p-value, degree of freedom and expected values.
    expected_values=values[3]
    degree_of_freedom=(len(observed_values[0,0:2])-1)*(len(observed_values[0:2,0])-1)
    print('Degree of freedom: ',degree_of_freedom)
    #calculating chi_sqaure value
    chi_square=sum([(obv-exp)**2/exp for obv,exp in zip(observed_values,expected_values)])
    print('Chi square: ',chi_square)
    chi_square_statistic=chi_square[0]+chi_square[1]
    print('Chi square statistic: ',chi_square_statistic)
    critical_value=chi2.ppf(q=1-alpha,df=degree_of_freedom)
    print(critical_value)
    #finding p-value.
    p_value=1-chi2.cdf(x=chi_square_statistic,df=degree_of_freedom)
    print('p_value:',p_value)
    if p_value<=alpha:
        print(h0)
    else:
        print(h1)
```

```python
for col1 in data_frame.columns:
    if col2 in data_frame.columns:
        if col1!=col2:
            if (data_frame[col1].dtype=='O')&(data_frame[col2].dtype=='O'):
                print('Chi2 test for {} and {}'.format(col1,col2))
                chi2_test(data_frame[col1],data_frame[col2])
                print('------------------------------------')
```
