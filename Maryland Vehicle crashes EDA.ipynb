## Null values

```python
null_percentage=[]
null_columns=[]
null_count=[]
for cols in data_frame.columns:
    total_null=data_frame[cols].isnull().sum()
    percentage=(total_null/len(data_frame))*100
    if percentage>0:
        null_percentage.append(percentage)
        null_columns.append(cols)
        null_count.append(total_null)
```

```python
pd.DataFrame({'Percentage':null_percentage,'Count':null_count},index=null_columns)
```

## Spliting the data frame into categorical and numerical features

```python
categorical_features=[]
numerical_features=[]
for cols in data_frame.columns:
    if data_frame[cols].dtype=='O':
        categorical_features.append(cols)
    else:
        numerical_features.append(cols)
```

## Visualizing the categorical features.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for cols in categorical_features:
    print('{} has {} unique features'.format(cols,len(data_frame[cols].unique())))
```

```python
for cols in categorical_features:
    if len(data_frame[cols].unique())<100:
        plt.figure(figsize=(10,10))
        sns.countplot(data_frame[cols])
        plt.xticks(rotation=90)
        plt.show()
```

## Visualizing the numerical features

```python
for cols in numerical_features:
    print('{} has {} unique values'.format(cols,len(data_frame[cols].unique())))
```

```python
for cols in numerical_features:
    if len(data_frame[cols].unique())<100:
        try:
            sns.countplot(data_frame[cols]);
            plt.xticks(rotation=90)
            plt.show()
        except:
            pass
```

## Removing the unwanted features

```python
for cols in data_frame.columns:
    null_count=data_frame[cols].isnull().sum()
    percentage=(null_count/len(data_frame))*100
    if percentage>=60:
        data_frame.drop([cols],axis=1,inplace=True)
```

## Replacing the missing values for unique values less than 100 with 'missing' term

```python
new_cols=['YEAR', 'QUARTER', 'LIGHT_DESC', 'COUNTY_DESC', 'JUNCTION_DESC', 'COLLISION_TYPE_DESC', 
'SURF_COND_DESC', 'LANE_DESC', 'RD_COND_DESC', 'FIX_OBJ_DESC', 'REPORT_TYPE', 'WEATHER_DESC', 'ACC_DATE', 
'ACC_TIME', 'SIGNAL_FLAG_DESC', 'HARM_EVENT_DESC2', 'RTE_NO', 'ROUTE_TYPE_CODE', 
'LOGMILE_DIR_FLAG', 'MAINROAD_NAME', 'FEET_MILES_FLAG', 
'DISTANCE_DIR_FLAG', 'REFERENCE_TYPE_CODE', 'REFERENCE_ROAD_NAME', 'LATITUDE', 'LONGITUDE', 'LOCATION']
```

```python
new_data=data_frame[new_cols]
```

```python
new_data.head()
```

```python
for cols in new_data.columns:
    if new_data[cols].dtype=='O':
        new_data[cols].fillna('Missing',inplace=True)
```

```python
years=[]
months=[]
dates=[]
for i in new_data['ACC_DATE']:
    i=str(i)
    year=i[:4]
    month=i[5:6]
    date=i[6:]
    years.append(year)
    months.append(month)
    dates.append(date)
```
